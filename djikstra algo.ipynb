# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {} # To store the path for reconstruction
    visited = []

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None # Initialize predecessors

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None: # All remaining nodes are unreachable or graph is disconnected
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node # Update predecessor

    return distances, predecessors
```

```python
# Helper function to reconstruct the shortest path
def reconstruct_path(predecessors, start_node, target_node):
    if target_node not in predecessors or (predecessors[target_node] is None and target_node != start_node):
        return [] # Target is unreachable or not in the graph

    path = [target_node]
    current_node = target_node

    # Trace back from target to start using predecessors
    while current_node != start_node:
        current_node = predecessors[current_node]
        if current_node is None: # This implies an issue or an unreachable path if target is not start
            return []
        path.insert(0, current_node)
    return path

# --- User-defined graph input ---
# You can modify this dictionary to represent your own graph.
# Each key is a node, and its value is another dictionary
# where keys are neighboring nodes and values are edge weights.
graph = {
    'A': {'B': 9, 'C': 2},
    'B': {'A': 9, 'C': 5, 'D': 10, 'E': 6}, # Added edge to E
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3, 'E': 7, 'F': 10}, # Added edges to E, F
    'E': {'B': 6, 'D': 7, 'F': 1}, # New node
    'F': {'D': 10, 'E': 1}  # New node
}

print("Testing Dijkstra with source node 'A':")
distances_A, predecessors_A = dijkstra(graph, 'A')

for node in distances_A:
    path = reconstruct_path(predecessors_A, 'A', node)
    if path:
        print(f"Distance from A to {node} = {distances_A[node]}, Path: {' -> '.join(path)}")
    else:
        print(f"Node {node} is unreachable from A.")

print("\nTesting Dijkstra with source node 'E':")
distances_E, predecessors_E = dijkstra(graph, 'E')

for node in distances_E:
    path = reconstruct_path(predecessors_E, 'E', node)
    if path:
        print(f"Distance from E to {node} = {distances_E[node]}, Path: {' -> '.join(path)}")
    else:
        print(f"Node {node} is unreachable from E.")
```

### Define Your Graph Interactively

To input your graph, please follow these steps:

1.  Enter the nodes of your graph, separated by spaces (e.g., `A B C D`).

3.  Finally, enter the starting node for Dijkstra's algorithm.

```python
def get_user_graph():
    nodes_input = input("Enter nodes separated by spaces (e.g., A B C D): ")
    nodes = nodes_input.split()

    user_graph = {}
    for node in nodes:
        user_graph[node] = {}
        neighbors_input = input(f"Enter neighbors and weights for node {node} (e.g., B:9 C:2, press Enter if no neighbors): ")
        if neighbors_input:
            neighbor_pairs = neighbors_input.split()
            for pair in neighbor_pairs:
                try:
                    neighbor, weight_str = pair.split(':')
                    weight = int(weight_str)
                    user_graph[node][neighbor] = weight
                except ValueError:
                    print(f"Invalid format for {pair}. Please use 'Neighbor:Weight'. Skipping this entry.")
    return user_graph

graph = get_user_graph()
start_node_input = input("Enter the starting node: ")
```
