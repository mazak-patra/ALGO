# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
def count_greater_than_zero(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values_linear = []

for n in n_values:
    # Create a list of 'n' elements, some positive, some not
    data = list(range(-n // 2, n // 2)) + [1] * (n - (n // 2 + n // 2))
    count_greater_than_zero(data)
    time_values_linear.append(n) # Time complexity is proportional to n

plt.figure()
plt.plot(n_values, time_values_linear)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for Counting Positive Elements")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
```

```python
import math

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0

    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps # Target not found, return total steps

n_values = list(range(1, 1001, 10)) # Input sizes from 1 to 1000
time_values_log = []

for n in n_values:
    # Create a sorted list for binary search
    data = list(range(n))
    # We'll search for the last element, which typically takes the most steps
    if n > 0:
        steps = binary_search(data, n - 1)
    else:
        steps = 0
    time_values_log.append(steps)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values_log, label='Binary Search Steps')
# Plot a theoretical log(n) curve for comparison
plt.plot(n_values, [math.log2(n) if n > 0 else 0 for n in n_values], linestyle='--', color='red', label='log2(n) (Theoretical)')
plt.xlabel("Input size n")
plt.ylabel("Number of Operations (Time)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.legend()
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
```

```python
def merge_sort(arr):
    steps = 0
    if len(arr) <= 1:
        return steps, arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    # Recursively sort the two halves
    left_steps, left = merge_sort(left)
    right_steps, right = merge_sort(right)
    steps += left_steps + right_steps

    # Merge the sorted halves
    i = j = k = 0
    while i < len(left) and j < len(right):
        steps += 1 # Counting comparison as a step
        if left[i] < right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    while i < len(left):
        steps += 1
        arr[k] = left[i]
        i += 1
        k += 1

    while j < len(right):
        steps += 1
        arr[k] = right[j]
        j += 1
        k += 1

    return steps, arr

n_values = list(range(1, 201, 5)) # Input sizes from 1 to 200, in steps of 5
time_values_n_log_n = []

for n in n_values:
    # Create a list of 'n' random elements for sorting
    data = list(range(n, 0, -1)) # Worst-case for some sorts, good for demonstration
    steps, _ = merge_sort(data)
    time_values_n_log_n.append(steps)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values_n_log_n, label='Merge Sort Steps')
# Plot a theoretical n*log(n) curve for comparison
import numpy as np
plt.plot(n_values, [x * np.log2(x) if x > 0 else 0 for x in n_values], linestyle='--', color='red', label='n * log2(n) (Theoretical)')
plt.xlabel("Input size n")
plt.ylabel("Number of Operations (Time)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.legend()
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in polynomial time. plot the graph as well, similar to above example
```

```python
def polynomial_time_example(n):
    steps = 0
    for i in range(n):
        for j in range(n):
            steps += 1  # This operation runs n*n times
    return steps

n_values = list(range(1, 101, 2)) # Input sizes from 1 to 100
time_values_polynomial = []

for n in n_values:
    steps = polynomial_time_example(n)
    time_values_polynomial.append(steps)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values_polynomial, label='O(n^2) Operations')
# Plot a theoretical n^2 curve for comparison
plt.plot(n_values, [n**2 for n in n_values], linestyle='--', color='red', label='n^2 (Theoretical)')
plt.xlabel("Input size n")
plt.ylabel("Number of Operations (Time)")
plt.title("O(n^2) Polynomial Time Complexity Example")
plt.legend()
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in exponential time. plot the graph as well, similar to above example
```

```python
def fibonacci_recursive(n):
    steps = 0
    if n <= 1:
        steps += 1 # Base case operation
        return steps, n

    # Each call generates two more calls, leading to exponential growth
    steps1, fib1 = fibonacci_recursive(n - 1)
    steps2, fib2 = fibonacci_recursive(n - 2)
    steps += steps1 + steps2 + 1 # Add 1 for the addition operation
    return steps, fib1 + fib2

n_values = list(range(1, 15)) # Exponential functions grow very quickly, so we use a small range
time_values_exponential = []

for n in n_values:
    steps, _ = fibonacci_recursive(n)
    time_values_exponential.append(steps)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values_exponential, label='Recursive Fibonacci Steps')
# Plot a theoretical 2^n curve for comparison (approximation for Fibonacci)
import numpy as np
plt.plot(n_values, [2**x for x in n_values], linestyle='--', color='red', label='2^n (Theoretical Upper Bound)')
plt.xlabel("Input size n")
plt.ylabel("Number of Operations (Time)")
plt.title("O(2^n) Exponential Time Complexity (Recursive Fibonacci)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i] # Swap
    arr[i + 1], arr[high] = arr[high], arr[i + 1] # Swap pivot
    return i + 1

def quick_sort_recursive_depth(arr, low, high, current_depth, max_depth):
    max_depth[0] = max(max_depth[0], current_depth) # Update max depth reached

    if low < high:
        pi = partition(arr, low, high)
        quick_sort_recursive_depth(arr, low, pi - 1, current_depth + 1, max_depth)
        quick_sort_recursive_depth(arr, pi + 1, high, current_depth + 1, max_depth)

def get_quick_sort_recursion_depth(arr):
    # Make a copy to avoid modifying the original array during depth calculation
    temp_arr = arr[:]
    max_depth = [0] # Use a list to pass by reference and track maximum depth
    quick_sort_recursive_depth(temp_arr, 0, len(temp_arr) - 1, 1, max_depth) # Start with depth 1
    return max_depth[0]

n_values = list(range(1, 201, 5)) # Input sizes from 1 to 200, in steps of 5
recursion_depth_values = []

for n in n_values:
    # Create a list of 'n' random elements for testing Quick Sort
    data = [random.randint(0, 1000) for _ in range(n)]
    depth = get_quick_sort_recursion_depth(data)
    recursion_depth_values.append(depth)

plt.figure(figsize=(10, 6))
plt.plot(n_values, recursion_depth_values, label='Quick Sort Recursion Depth')
# Plot a theoretical log(n) curve for comparison
plt.plot(n_values, [np.log2(x) if x > 0 else 0 for x in n_values], linestyle='--', color='red', label='log2(n) (Theoretical)')
plt.xlabel("Input size n")
plt.ylabel("Maximum Recursion Depth")
plt.title("O(log n) Aspect of Quick Sort (Recursion Depth)")
plt.legend()
plt.grid(True)
plt.show()

print("\nNote: The plot above illustrates the O(log n) recursion depth of Quick Sort in the average case.")
print("However, the *overall time complexity* for sorting `n` elements using Quick Sort is O(n log n) on average.")
print("In the worst case (e.g., already sorted array with a poor pivot choice), the recursion depth can be O(n),")
print("leading to an overall time complexity of O(n^2).")
```
